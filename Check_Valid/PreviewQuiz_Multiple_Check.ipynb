## 데이터 전처리

```python
import pandas as pd

df_mid = pd.read_csv("DB_MID.csv")
original_mid_data = df_mid.copy()
df_mid_check = pd.read_csv("multiple_check_mid.csv")
```

```python
df_mid.head(1)
```

```python
df_mid.columns
```

```python
df_mid.columns
drop_col = ['textbook ID', '학년', '출판사', '교육과정', 'Lesson', '구분',
       'title', 'DESCRIPTION',
       'Key Expression (한국어)','Sub Expression (한국어)',
       '학습목표', 'multiple check', 'Quiz 1', 'Korean 1', 'Option 1-1',
       'Option 1-2', 'Option 1-3', 'Answer 1', 'Index 1', 'Quiz 2', 'Korean 2',
       'Option 2-1', 'Option 2-2', 'Option 2-3', 'Answer 2', 'Index 2',
       '샘플 스크립트 (영어)', '샘플 스크립트 (한국어)', 'Unnamed: 32',
       'STRUCTURE - Title, New Desc 작업용', 'Unnamed: 34', 'Unnamed: 35',
       'Unnamed: 36']
df_mid.drop(columns=drop_col, inplace=True)
df_mid = df_mid.rename(columns={
    "curriculum id" : "id",
    "new_DES" : "description",
    "Key Expression (영어)" : "key_expression",
    "Sub Expression (영어)" : "sub_expression"
})
df_mid.head(1)
```

```python
df_mid_check.head(1)
```

```python
df_mid_check.columns
drop_col = ['possible_answer.0',
       'possible_answer.2', 'reason', "description", "key expression", "sub expression"]
df_mid_check.drop(columns=drop_col, inplace=True)
print(len(df_mid_check))
df_mid_check.head(1)
```

```python
df_mid_check['possible_answer.1'].value_counts().sum()
```

```python
df_mid.head()
```

## 조인 결과

```python
merged_data_mid = pd.merge(df_mid_check, df_mid, on="id")
merged_data_mid.head(3)
#merged_data_mid["description"].isna()
```

```python
display(merged_data_mid_quiz1.head(3))
display(merged_data_mid_quiz2.head(3))
```

```python
display(merged_data_mid_quiz1.head(3))
display(merged_data_mid_quiz2.head(3))
```

```python
merged_data_mid_quiz1.to_csv("PreviewQuiz1_MID.csv", encoding="utf-8-sig", index=False)
merged_data_mid_quiz2.to_csv("PreviewQuiz2_MID.csv", encoding="utf-8-sig", index=False)
```

## 결과값 불러와서 업데이트 하기

```python
# 결과 csv 파일 불러오기
result_mid_quiz1 = pd.read_csv("PreviewQuiz1_MID_Multiple_Check_Result.csv")
result_mid_quiz2 = pd.read_csv("PreviewQuiz2_MID_Multiple_Check_Result.csv")

result_mid_quiz2.head(3)
```

```python
# 데이터 전처리
drop_col = ["result_json", "id.1"]

result_mid_quiz1.drop(columns=drop_col, inplace=True)
result_mid_quiz1 = result_mid_quiz1.rename(columns={
    "id" : "curriculum id",
    "fill_in_blank_quiz" : "Quiz 1",
    "korean_translation" : "Korean 1",
    "option_1" : "Option 1-1",
    "option_2" : "Option 1-2",
    "option_3" : "Option 1-3",
    "answer" : "Answer 1",
    "answer_index" : "Index 1"
})

result_mid_quiz2.drop(columns=drop_col, inplace=True)

result_mid_quiz2 = result_mid_quiz2.rename(columns={
    "id" : "curriculum id",
    "fill_in_blank_quiz" : "Quiz 2",
    "korean_translation" : "Korean 2",
    "option_1" : "Option 2-1",
    "option_2" : "Option 2-2",
    "option_3" : "Option 2-3",
    "answer" : "Answer 2",
    "answer_index" : "Index 2"
})
```

```python
result_mid_quiz1.head(5)
```

```python
result_mid_quiz1.columns
```

```python
result_mid_quiz2.head(5)
```

```python
original_mid_data.columns
```

```python
# 업데이트할 컬럼 목록
columns_to_update_quiz1 = ['Quiz 1', 'Korean 1', 'Option 1-1', 'Option 1-2', 'Option 1-3', 'Answer 1', 'Index 1']
columns_to_update_quiz2 = ['Quiz 2', 'Korean 2', 'Option 2-1', 'Option 2-2', 'Option 2-3', 'Answer 2', 'Index 2']

# 'curriculum id'를 인덱스로 설정 (더 정확한 매칭을 위해)
original_mid_data.set_index('curriculum id', inplace=True)
result_mid_quiz1.set_index('curriculum id', inplace=True)
result_mid_quiz2.set_index('curriculum id', inplace=True)

# 바로 값 덮어쓰기 (NaN 여부 상관없이 무조건 덮어씀)
original_mid_data.update(result_mid_quiz1[columns_to_update_quiz1])
original_mid_data.update(result_mid_quiz2[columns_to_update_quiz2])

# 인덱스를 다시 컬럼으로 복구
original_mid_data.reset_index(inplace=True)

# CSV 저장
original_mid_data.to_csv("PreviewQuiz_MID_Multiple_Check_Result.csv", encoding="utf-8-sig", index=False)

print("업데이트 완료! CSV 파일이 저장되었습니다.")
```
