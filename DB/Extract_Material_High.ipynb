```python
def split_material(material):
    result = {
        'DESCRIPTION': '',
        'KEY_EXPRESSION': '',
        'KEY_EXPRESSION_KOR': '',
        'SUB_EXPRESSION': ''
    }
    
    current = None
    for line in material:
        line = line.strip()
    
        if '[DESCRIPTION]' in line:
            current = 'DESCRIPTION'
        elif '[KEY_EXPRESSION]' in line:
            current = 'KEY_EXPRESSION'
        elif '[KEY_EXPRESSION_KOR]' in line:
            current = 'KEY_EXPRESSION_KOR'
        elif '[SUB_EXPRESSION]' in line:
            current = 'SUB_EXPRESSION'       
        elif '[SUB_EXPRESSION_1]' in line:
            current = 'SUB_EXPRESSION'
        elif '[SUB_EXPRESSION_2]' in line:
            current = 'SUB_EXPRESSION'            
        elif line:  # 빈 줄이 아니면 현재 항목에 내용 추가
            if current:
                result[current] += line + ' '

    return result
```

```python
df['DESCRIPTION'] = ''
df['KEY_EXPRESSION'] = ''
df['KEY_EXPRESSION_KOR'] = ''
df['SUB_EXPRESSION'] = ''

for i, row in df.iterrows():
    raw = split_material(description[i])
    df.at[i, 'DESCRIPTION'] = raw['DESCRIPTION']
    df.at[i, 'KEY_EXPRESSION'] = raw['KEY_EXPRESSION']
    df.at[i, 'KEY_EXPRESSION_KOR'] = raw['KEY_EXPRESSION_KOR']
    df.at[i, 'SUB_EXPRESSION'] = raw['SUB_EXPRESSION']

df.head(3)
```
