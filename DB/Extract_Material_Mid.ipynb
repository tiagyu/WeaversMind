```python
def split_material(material):
    result = {
        'DESCRIPTION': '',
        'KEY_EXPRESSION': '',
        'KEY_EXPRESSION_KOR': '',
        'SUB_EXPRESSION': ''
    }
    
    current = None
    for line in material:
        line = line.strip()
    
        if '[DESCRIPTION]' in line:
            current = 'DESCRIPTION'
        elif '[KEY_EXPRESSION]' in line:
            current = 'KEY_EXPRESSION'
        elif '[KEY_EXPRESSION_KOR]' in line:
            current = 'KEY_EXPRESSION_KOR'
        elif '[SUB_EXPRESSION]' in line:
            current = 'SUB_EXPRESSION'
        elif line:  # 빈 줄이 아니면 현재 항목에 내용 추가
            if current:
                result[current] += line + ' '

    return result
```

```python
df['DESCRIPTION'] = ''
df['KEY_EXPRESSION'] = ''
df['KEY_EXPRESSION_KOR'] = ''
df['SUB_EXPRESSION'] = ''

for i, row in df.iterrows():
    raw = split_material(description[i])
    df.at[i, 'DESCRIPTION'] = raw['DESCRIPTION']
    df.at[i, 'KEY_EXPRESSION'] = raw['KEY_EXPRESSION']
    df.at[i, 'KEY_EXPRESSION_KOR'] = raw['KEY_EXPRESSION_KOR']
    df.at[i, 'SUB_EXPRESSION'] = raw['SUB_EXPRESSION']

df.head(3)
```

## TEXT ID

```python
id = pd.read_csv('Text_ID.csv')
id.columns
drop_col = ['구분','교과서 확보', '어휘 자료','function', 'structure', 'Issue']
id.drop(columns=drop_col, inplace=True)
id['학년 정보'] = id['학년 정보'].astype(str).replace('영어', '', regex=True)
id.rename(columns={
    '학년 정보': '학년',
    '개정 정보': '교육과정'
}, inplace=True)
display(id)
```

```python
# 변환 함수 정의
def modify_publisher(name):
    if "(" in name and ")" in name:  # 저자 이름이 있는 경우만 처리
        parts = name.split("(")  # 괄호를 기준으로 분리
        publisher = parts[0].strip()  # 출판사 이름
        author_initial = parts[1][0]  # 저자의 첫 글자만 추출
        return f"{publisher}({author_initial})"
    return name  # 저자가 없는 경우 그대로 반환

# 데이터 변환 적용
id["출판사"] = id["출판사"].apply(modify_publisher)
```

```python
id['출판사'] = id['출판사'].replace('비상(김)','비상')
```
