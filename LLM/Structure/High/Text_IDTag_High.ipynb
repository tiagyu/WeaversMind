# 새로운 칼럼 생성

```python
df['출판사'] = df['출판사'].replace("비상", "비상(홍)")
df['제목'] = df['제목'].replace("영어1", "영어 1")
```

```python
df['common columns'] = df['교육과정'].astype(str) + "-" + df['제목'].astype(str) + "-" + df['출판사'].astype(str)
df.head(1)
```

# 포맷팅 작업

```python
type(id['교육과정'][0])
```

```python
def formatting(subject:str, publisher:str, year:int) -> str:
    if "(" in subject and ")" in subject: # 저자 이름이 있는 경우만 처리
        parts = subject.split("(") # 괄호 기준으로 나눔
        subject_format = parts[0].strip() # 제목 이름
        author_name = "(" + parts[1][0] + ")"
        publisher_format = publisher + author_name
        return f"{str(year)}-{subject_format}-{publisher_format}"
    return subject, publisher
```

```python
id['common columns'] = id.apply(lambda row: formatting(row['제목'], row["출판사"], row['교육과정']), axis=1)
```

```python
id.head()
```

```python
df['출판사'].value_counts()
```

```python
merged_df = df.merge(id[['textbook ID','common columns']], on = 'common columns', how = 'left')
merged_df.head(1)
```

```python
merged_df['textbook ID'] = merged_df['textbook ID'].astype("Int64")
# merged_df['textbook ID'].value_counts()
merged_df.isna().sum()
```

```python
na_textbook_id_rows = merged_df[merged_df["textbook ID"].isna()]

# 결과 출력
print(na_textbook_id_rows)
```
